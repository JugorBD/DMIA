```python
import pandas as pd
import numpy as np

from sklearn.ensemble import GradientBoostingClassifier
```

Читаем входные файлы с данными

```python
transactions = pd.read_csv('transactions.csv')
customers_gender = pd.read_csv('customers_gender_train.csv')
tr_mcc_code = pd.read_csv('tr_mcc_codes.csv')
tr_type = pd.read_csv('tr_types.csv')
```

```python
transactions.mcc_code.unique()
```

```python
for customer in sorted(transactions.customer_id.unique()):
    df_customer = transactions[transactions.customer_id==customer]
    for mcc_code in df_customer.mcc_code.unique():
        df_customer_mcc_code = df_customer[df_customer.mcc_code==mcc_code]
        break
    break
df_customer_mcc_code.amount.sum()
```

```python
Z = transactions.groupby('customer_id').apply(lambda x: x[['amount']].mean())
```

```python
Z.head(10)
```

```python
tr_mcc_code.head(5)
```

Для каждого клиента подсчитываем различные mcc_code

```python
X = transactions.groupby('customer_id') \
                    .apply(lambda x: x[['mcc_code', 'tr_type']].unstack().value_counts()) \
                    .unstack() \
                    .fillna(0)
```

```python
X = pd.concat([Z, X], axis=1)
```

```python
X = pd.DataFrame(index=sorted(transactions.customer_id.unique()), columns=list(transactions.mcc_code.unique()) \
                                                                            +   list(transactions.tr_type.unique())).fillna(0)
```

```python
X.head()
```

```python
for customer in sorted(transactions.customer_id.unique()):
    df_customer = transactions[transactions.customer_id==customer]
    for mcc_code in df_customer.mcc_code.unique():
        df_customer_mcc_code = df_customer[df_customer.mcc_code==mcc_code]
        X.loc[[customer], [mcc_code]] = df_customer_mcc_code.amount.sum()
        
```

```python
for customer in sorted(transactions.customer_id.unique()):
    df_customer = transactions[transactions.customer_id==customer]
    for tr_type in df_customer.tr_type.unique():
        df_customer_tr_type = df_customer[df_customer.tr_type==tr_type]
        X.loc[[customer], [tr_type]] = df_customer_tr_type.amount.sum()
```

```python
X.head(50)
```

```python
customers_gender = customers_gender.set_index('customer_id')
```

```python
customers_gender.head()
```

```python
Y_train = customers_gender.loc[X.index].gender
Y_train = Y_train.reset_index()
del Y_train['customer_id']
Y_train = Y_train.dropna(0)
```

```python
Y_train.head()
```

```python
X_train = X.reset_index()
X_train = X_train.loc[Y_train.index].set_index('customer_id')
```

```python
X_train.head()
```

Обучаемся на всех данных

```python
clf = GradientBoostingClassifier(random_state=13)
clf.fit(X_train, Y_train.values[:, 0])
```

Предсказываем пол для тестовых данных и создаём файл с ответом

```python
X_test = X.drop(customers_gender.index)
result = pd.DataFrame(X_test.index, columns=['customer_id'])
result['gender'] = clf.predict_proba(X_test)[:, 1]
```

```python
result.to_csv('baseline_a.csv', index=False)
```
